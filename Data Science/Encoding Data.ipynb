```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder 
```

```python
df = sns.load_dataset("titanic")
```

```python
df 
```

```python
df.survived.value_counts()
```

```python
df[["sex", "class", "embark_town"]].isnull().sum()
```

```python
df = df.dropna(subset = ["embark_town"])
```

```python
df[["sex", "class", "embark_town"]].isnull().sum()
```

```python
df.sex.value_counts()
```

```python
df["class"].value_counts()
```

```python
df["embark_town"].value_counts()
```

```python
df_onehot = pd.get_dummies(df, columns = ["sex", "embark_town"], drop_first = True)
```

```python
df_onehot.columns
```

```python
df_onehot
```

```python
#Â label encoder 
```

```python
label_encoder = LabelEncoder()
```

```python
df_label = df.copy()
```

```python
df_label 
```

```python
df_label["sex"] = label_encoder.fit_transform(df_label["sex"])
```

```python
df_label
```

```python
#Â ordinal encoder 
```

```python
df_ordinal = df.copy()
```

```python
ordinal_encoder = OrdinalEncoder(categories = [["Third", "Second", "First"]])
```

```python
df_ordinal["class"] = ordinal_encoder.fit_transform(df_ordinal[["class"]])
```

```python
df_ordinal
```

```python
fig, axes = plt.subplots(1,3, figsize = (15,5))
df.sex.value_counts().plot(kind = "bar", ax = axes[0], title = "Original Categorical")
df_label.sex.value_counts().plot(kind = "bar", ax = axes[1], title = "Label Encoded")
df_onehot.sex_male.value_counts().plot(kind = "bar", ax = axes[2], title = "Onehot Encoded")
```

```python
df_ordinal["class"].value_counts().plot(kind = "bar", title = "Ordinal Encoded")
```
