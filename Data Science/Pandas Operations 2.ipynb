```python
def salary_category(salary): 
    if salary < 50000: 
        return "Low"
    elif 50000 < salary < 80000: 
        return "Medium"
    else: 
        return "High"
```

```python
dataFrame["Salary"].apply(salary_category)
```

```python
dataFrame["Salary_Category"] = dataFrame["Salary"].apply(salary_category) 
```

```python
dataFrame
```

```python
def newPerformanceScore(experience): 
    if experience > 10: 
        return 1
    return 0
```

```python
dataFrame["New_Score"] = dataFrame["Performance_Score"] + dataFrame["Experience"].apply(newPerformanceScore) 
```

```python
dataFrame
```

```python
def adjust_new(row): 
    if row["Experience"] > 10: 
        return row["Performance_Score"] + 1
    else: 
        return row["Performance_Score"]
```

```python
dataFrame["Adjusted_Score"] = dataFrame.apply(adjust_new, axis = 1)
```

```python
dataFrame
```

```python
dataFrame["Formatted_Name"] = dataFrame["Name"].apply(lambda x: x.replace("_", " "))
```

```python
dataFrame
```

```python
dataFrame.drop("Adjusted_Score", axis = 1, inplace = True)
```

```python
dataFrame
```

```python
dataFrame.drop("Formatted_Name", axis = 1, inplace = True)
```

```python
dataFrame
```

```python
def changeDepartment(row): 
    if row["Experience"] > 5 and (row["Department"] == 'Marketing' or row["Department"] == 'Sales'): 
        row["Department"] = 'Finance'
    return row
```

```python
dataFrame.apply(changeDepartment, axis = 1)
```
